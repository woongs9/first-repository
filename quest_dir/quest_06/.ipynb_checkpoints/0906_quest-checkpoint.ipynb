```python
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

import pandas as pd
```

```python
def runing_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(model.__class__.__name__)
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print('------------------------------------------------')
```

```python
model_list = [
    DecisionTreeClassifier(random_state = 2),
    RandomForestClassifier(random_state = 0),
    svm.SVC(random_state = 0),
    SGDClassifier(random_state = 166),
    LogisticRegression(random_state = 32, max_iter = 5000)
]
```

# Load digit : 손글씨 분류

```python
digits = load_digits()
print(digits.keys())
```

```python
# Feature Data 지정하기
digits_data = digits.data

# Label Data 지정하기
digits_label = digits.target

digits_data.shape, digits_label.shape
```

```python
# Target Names 출력해 보기
digits.target_names
```

```python
# 데이터 Describe 해 보기
digits.DESCR
```

```python
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
digits_df['Label'] = digits.target
digits_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                   digits_label,
                                                   test_size = 0.2,
                                                   random_state = 7)
```

```python
for model in model_list:
    runing_model(model, X_train, X_test, y_train, y_test)
```

## 모델 평가하기

digits 파일은 라벨 데이터가 어느 한쪽으로 치우쳐지지 않았기 때문에 accuracy 점수로 평가를 해도 좋을 것 같다.

# load_breast_cancer : 유방암 여부 진단

```python
breast_cancer = load_breast_cancer()
breast_cancer.keys()
```

```python
# Feature Data 지정하기
breast_cancer_data = breast_cancer.data

# Label Data 지정하기
breast_cancer_label = breast_cancer.target

breast_cancer_data.shape, breast_cancer_label.shape
```

```python
# Target Names 출력해 보기
breast_cancer.target_names
```

```python
# 데이터 Describe 해 보기
breast_cancer.DESCR
```

```python
breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names)
breast_cancer_df['Label'] = breast_cancer.target
breast_cancer_df
```

```python
X_breast_cancer_train, X_breast_cancer_test, y_breast_cancer_train, y_breast_cancer_test = train_test_split(breast_cancer_data,
                                                   breast_cancer_label,
                                                   test_size = 0.2,
                                                   random_state = 7)
```

## 높은 점수의 Randum_state 구해보기

```python
lis = []
for i in range(100):
    di = DecisionTreeClassifier(random_state=i)
    di.fit(X_breast_cancer_train, y_breast_cancer_train)
    y_pred = di.predict(X_breast_cancer_test)
    accuracy = accuracy_score(y_breast_cancer_test, y_pred)
    lis.append(accuracy)

print(lis.index(max(lis)))
```

```python
lis = []
for i in range(10):
    di = RandomForestClassifier(random_state=i)
    di.fit(X_breast_cancer_train, y_breast_cancer_train)
    y_pred = di.predict(X_breast_cancer_test)
    accuracy = accuracy_score(y_breast_cancer_test, y_pred)
    lis.append(accuracy)

print(lis.index(max(lis)))
```

```python
lis = []
for i in range(100):
    di = svm.SVC(random_state=i)
    di.fit(X_breast_cancer_train, y_breast_cancer_train)
    y_pred = di.predict(X_breast_cancer_test)
    accuracy = accuracy_score(y_breast_cancer_test, y_pred)
    lis.append(accuracy)

print(lis.index(max(lis)))
```

```python
lis = []
for i in range(100):
    di = SGDClassifier(random_state=i)
    di.fit(X_breast_cancer_train, y_breast_cancer_train)
    y_pred = di.predict(X_breast_cancer_test)
    accuracy = accuracy_score(y_breast_cancer_test, y_pred)
    lis.append(accuracy)

print(lis.index(max(lis)))
```

```python
lis = []
for i in range(100):
    di = LogisticRegression(random_state=i, max_iter=5000)
    di.fit(X_breast_cancer_train, y_breast_cancer_train)
    y_pred = di.predict(X_breast_cancer_test)
    accuracy = accuracy_score(y_breast_cancer_test, y_pred)
    lis.append(accuracy)

print(lis.index(max(lis)))
```

```python
for model in model_list:
    runing_model(model, X_breast_cancer_train, X_breast_cancer_test, y_breast_cancer_train, y_breast_cancer_test)
```

## 모델 평가하기

유방암환자 한명이라도 양성을 음성으로 판단하면 안되기 떄문에 Recall 점수로 평가를 해야한다.

# 회고

## 이번 프로젝트를 하면서 어려웠던 점 ##
- 이번 프로젝트는 지난 프로젝트를 진행하면서 익혔던 내용이라 딱히 어려움 없이 잘 진행할 수 있었다.

## 알아낸 점 ##
- 이번 프로젝트를 진행하면서 반복되는 코드를 함수화 하여 간소화 할수 있다는 것을 알고는 있었지만 실제로 코딩을 해 보면서 좀더 익숙 해 진것 같다.
- 이번 프로젝트를 하면서 confusion_matrix를 출력 해 봤는데 2x2 행렬이 나오는 것이 아닌 digits 데이터는 10x10 행렬이 나오고 wine 데이터는 3x3, 유방암 데이터는 2x2가 나와 왜 그런지 몰라 당황했었다. 알고보니 label 값의 갯수만큼의 행렬이 생성된다는 것을 이해 하였다.
- 처음 모델의 random_state를 임의로 설정 후 실행하니 어느 모델에서는 특정 라벨값을 전혀 예측 하지 않아 경고가 발생하는 상황을 확인했다. 그래서 random_state를 수정 후 다시 실행하니 그 경고가 나오지 않는것을 확인했다. random_state를 임의로 결정하지 않고 각 모델별로 random_state별 accuracy 점수가 가장 높은 random_state를 구해보았다. 그 후 점수가 더 올라간 것을 확인했다.

## 모호한 점 ##
- 확실하지는 않지만 random_state는 말 그대로 random 함수의 시드 값이기 때문에 모델의 성능에 직접적인 연관이 없을 것 같다. 해서 시도만 해보고 모든 데이터에 적용하여 random_state를 찾아보지는 않았다.
